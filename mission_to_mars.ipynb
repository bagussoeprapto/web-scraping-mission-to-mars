```python
# Dependencies
from bs4 import BeautifulSoup
import requests
```

### NASA Mars News

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(url)

# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Retrieve the parent divs for all articles
results = soup.find_all('div', class_='slide')

# Loop through results to retrieve article title, header, and timestamp of article
for result in results:
    title = result.find('div', class_='content_title').text
    excerpt = result.find('div', class_='rollover_description_inner').text

    print('-----------------')
    print(title)
    print(excerpt)
```

### JPL Mars Space Images - Featured Image

```python
# Dependencies
from splinter import Browser
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url2 = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url2)
```

```python
html = browser.html
soup2 = BeautifulSoup(html, 'html.parser')
```

```python
one = soup2.find('article', class_='carousel_item')
img_url = one["style"]
img_url = img_url[23:75]
print('https://www.jpl.nasa.gov/' + img_url)
```

### Mars Hemispheres

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url5 = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url5)
```

```python
html = browser.html
soup4 = BeautifulSoup(html, 'html.parser')
```

```python
titles = soup4.find_all('img', class_='thumb')

for title in titles:
    img = title["src"]
    print('-----------------')
    print('https://astrogeology.usgs.gov/' + img)
```
